Creación e Información	

Importar la librería	                     import pandas as pd	      >>Importa Pandas, usualmente con el alias pd.
Crear DataFrame	                             df = pd.DataFrame(datos)	  >>Crea un DataFrame a partir de un diccionario o una lista de listas.
Cargar desde CSV	                         df = pd.read_csv('archivo.csv')	 >>Carga datos desde un archivo CSV.
Mostrar las primeras N filas	             df.head(n=5)	              >>Muestra las primeras 5 filas (o n filas).
Mostrar información general	                 df.info()	                  >>Muestra tipos de datos, conteo de no-nulos y uso de memoria.
Estadísticas descriptivas	                 df.describe()	              >> Muestra el resumen estadístico de las columnas numéricas.
Nombres de columnas	                         df.columns	                  >>Devuelve el objeto Index con los nombres de las columnas.

Selección de Datos	

Seleccionar una columna	                     df['Nombre_Columna']	Devuelve una Serie de Pandas.
Seleccionar múltiples columnas	             df[['Col_A', 'Col_B']]	Devuelve un DataFrame.
Seleccionar por índice (filas/columnas)    	df.iloc[fila, columna]	Selección por posición entera (ej: df.iloc[0:5, [1, 3]]).
Seleccionar por etiqueta	                df.loc[etiqueta_fila, 'Etiqueta_Columna']	Selección por etiqueta de fila y nombre de columna.

Filtrado	

Filtrar filas por condición	                df[df['Col_A'] > 10]	Devuelve las filas donde la condición es Verdadera.
Filtrar con múltiples condiciones	        df[(df['Col_A'] > 10) & (df['Col_B'] == 'pop')]	Uso de & (AND) u `

Manipulación

Añadir nueva columna	                   df['Nueva_Col'] = valores	Asigna una lista o Serie de valores a una nueva columna.
Eliminar columnas/filas	                   df.drop(columns=['Col_A'])	Elimina columnas (se usa axis=1 para columnas, axis=0 para filas).
Agrupar y agregar	                       df.groupby('Col_A')['Col_B'].mean()	Agrupa por 'Col_A' y calcula la media de 'Col_B'.
Ordenar valores	                           df.sort_values(by='Col_A', ascending=False)	Ordena el DataFrame por la columna especificada.

Valores Nulos (NaN)	

Identificar nulos	                       df.isnull().sum()	Cuenta el total de valores nulos por columna.
Eliminar filas con nulos	               df.dropna()	Elimina filas que contienen al menos un valor nulo.
Rellenar nulos	                           df.fillna(valor)	Rellena los valores nulos con un valor específico (ej: 0 o la media).
